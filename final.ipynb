```python
import pyspark
from operator import add
sc = pyspark.SparkContext('local[*]')
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
```

Ejecutando el motor de spark

```python
lines = sqlContext.read.text('noticiatulio.txt').rdd.map(lambda r: r[0])
counts = lines.flatMap(lambda x: x.split(' ')) \
.map(lambda x: (x, 1)) \
.reduceByKey(add)
```

Primero se lee el archivo de texto mapeando linea por linea
Luego divide cada linea por los espacios ' ' para obtener las palabras
despues mapea y reduce por las palabras comunes


```python
output = counts.collect()
for (word, count) in output:
    print("%s: %i" % (word, count))
```

Aqui ejecuta el codigo
luego recorre la variable output teniendo en cuenta las dos llaves del array
mientras las recorre las va mostrando
